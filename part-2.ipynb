```python
# Plot the bar graph top 15 restaurants have a maximum number of outlets.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
zomato_df =  pd.read_csv("zomato.csv",encoding='latin-1')
temp = zomato_df.copy()
zomato_df = zomato_df[zomato_df['Country Code']==1]
res_count = {}
for i in zomato_df['Restaurant Name'] :
    res_count[i] = res_count.get(i,0)+1
res_count = [[value,key] for key,value in res_count.items()]
res_count.sort(reverse=True)
restaurants_name = [ i[1] for i in res_count][:19]
counts = [ i[0] for i in res_count][:19]

plt.bar(restaurants_name,counts)
plt.xlabel("Restaurant", size=15)
plt.ylabel("No of Outlets",size = 15)
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the histogram of aggregate rating of restaurant( drop the unrated restaurant).
```

```python
import numpy as np
```

```python
new_df = zomato_df.copy()
new_df = zomato_df[zomato_df['Rating text']!='Not Rated'].copy()
ratings = np.array(new_df['Aggregate rating'])
real_ratings = ~(np.isnan(ratings))
ratings = ratings[real_ratings]
plt.hist(ratings,bins="auto")
plt.xticks(np.arange(6))
plt.show()
```

```python
# Plot the bar graph top 10 restaurants in the data with the highest number of votes
```

```python
name_votes = {}
for i,j in zip(zomato_df['Restaurant Name'],zomato_df['Votes']) :
    name = i
    vote = j
    name_votes[name] = name_votes.get(name,0) + vote
name_votes = [[value,key] for key,value in name_votes.items()]
name_votes.sort(reverse=True)
votes = [i[0] for i in name_votes][:10]
names = [i[1] for i in name_votes][:10]
plt.bar(names,votes)
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the pie graph of top 10 cuisines present in restaurants in the USA.
```

```python
zomato_usa_df = temp[temp['Country Code']==216].copy()
zomato_usa_df['Cuisines'].dropna(inplace=True)
cuisines_count_usa = {}
for i in zomato_usa_df['Cuisines'] :
    cuisines = str(i)
    if cuisines!="":
        for cuisine in cuisines.split(', '):
            cuisines_count_usa[cuisine] = cuisines_count_usa.get(cuisine,0)+1
cuisines_count = [[value,key] for key,value in cuisines_count_usa.items()]
cuisines_count.sort(reverse=True)
counts = [i[0] for i in cuisines_count][:10]
cuisine_names = [i[1] for i in cuisines_count][:10]
plt.pie(counts,labels=cuisine_names)
plt.show()
```
